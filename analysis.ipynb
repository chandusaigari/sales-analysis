```python
import pandas as pd      #changing datatypes to thier respective columns

df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='mixed', errors='coerce')
df['Size'] = df['Size'].astype('string')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce').astype('Int64')
df['Total_Amount'] = pd.to_numeric(df['Total_Amount'], errors='coerce')
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce').astype('Int64')
df['Feedback'] = df['Feedback'].astype('object')
df['City'] = df['City'].astype('object')
df['Delivery_Days'] = pd.to_numeric(df['Delivery_Days'], errors='coerce').astype('Int64')

print(df.dtypes)
```

```python
import pandas as pd      #mappping for space reduction

word_to_num = {
    'zero': 0, 'one': 1, 'two': 2, 'three': 3, 'four': 4,
    'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9, 'ten': 10
}

# convert words to numbers first
df['Price'] = df['Price'].replace(word_to_num)

# convert to numeric, invalid values -> NaN, then fill and convert to int
df['Price'] = pd.to_numeric(df['Price'], errors='coerce').fillna(0).astype(int)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Total_Amount'] = (
    df['Total_Amount']
    .astype(str)
    .str.replace(r'[^0-9.]', '', regex=True)
)

df['Total_Amount'] = pd.to_numeric(df['Total_Amount'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df['Total_Amount'] = df['Total_Amount'].astype(str).str.replace(r'[^\d.,-]', '', regex=True)#same as trim()
```

```python
Q1 = df['Total_Amount'].quantile(0.25)#for outlier finding 
Q3 = df['Total_Amount'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Total_Amount'] < lower_bound) | (df['Total_Amount'] > upper_bound)]

print(outliers)
```
